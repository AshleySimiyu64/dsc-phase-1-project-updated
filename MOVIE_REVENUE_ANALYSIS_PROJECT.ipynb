```python
import pandas as pd
import matplotlib.pyplot as plt

movie_gross_data =pd.read_csv('bom.movie_gross.csv')
movie_gross_data.head()
```

```python
title_basic_data =pd.read_csv('title.basics.csv')
title_basic_data.head()
```

```python
title_ratings_data =pd.read_csv('title.ratings.csv')
title_ratings_data.head()
```

```python
merged_file = pd.merge(title_ratings_data,title_basic_data,on='tconst')
merged_file.head()
```

```python
# Merging of data from movie_gross_data and first set of merged data
final_merged_data = pd.merge(movie_gross_data,merged_file, left_on=('title','year'),right_on=('primary_title','start_year'))
final_merged_data.head()
```

```python
# Drop rows with missing values in title or domestic_gross
movies_gross_domestic = final_merged_data.dropna(subset=['title', 'domestic_gross'])

# Group by movie title and sum domestic gross over the years
total_domestic = movies_gross_domestic.groupby('title')['domestic_gross'].sum().reset_index()

# Get top 10 movies by total domestic gross
top10 = total_domestic.sort_values(by='domestic_gross', ascending=False).head(10)
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(12, 6))
plt.barh(top10['title'], top10['domestic_gross'], color='steelblue')
plt.xlabel('Total Domestic Gross')
plt.title('Top 10 Movies by Domestic Gross (2010–2018)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Step 1: Drop rows with missing 'title' or 'domestic_gross'
movies_gross_domestic = final_merged_data.dropna(subset=['title', 'domestic_gross'])

# Step 2: Group by title and sum domestic gross
total_domestic = movies_gross_domestic.groupby('title', as_index=False)['domestic_gross'].sum()

# Step 3: Get top 10 movies by domestic gross
top10 = total_domestic.sort_values(by='domestic_gross', ascending=False).head(10)

# Step 4: Plot
plt.figure(figsize=(12, 6))
bars = plt.barh(top10['title'], top10['domestic_gross'], color='steelblue')

# Add value labels to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1e6, bar.get_y() + bar.get_height() / 2,
             f"${int(width):,}", va='center', fontsize=9)

# Final touches
plt.xlabel('Total Domestic Gross (USD)')
plt.title('Top 10 Movies by Domestic Gross (2010–2018)')
plt.gca().invert_yaxis()  # So highest grossing is on top
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data — replace with your actual data
df_top10 = pd.DataFrame({
    'movie_title': [
        'Star Wars: The Force Awakens', 'Avengers: Endgame', 'Black Panther',
        'Avengers: Infinity War', 'Jurassic World', 'The Lion King',
        'The Avengers', 'Incredibles 2', 'Beauty and the Beast', 'Frozen'
    ],
    'release_year': [2015, 2019, 2018, 2018, 2015, 2019, 2012, 2018, 2017, 2013],
    'domestic_gross': [936, 858, 700, 679, 652, 543, 623, 609, 504, 477]
})

# Plotting
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_top10, x='release_year', y='domestic_gross', marker='o', hue='movie_title', palette='tab10')

# Title and labels
plt.title('Top 10 Movie Gross Revenue by Release Year (2010–2018)', fontsize=14)
plt.xlabel('Release Year')
plt.ylabel('Domestic Gross (in Millions USD)')
plt.xticks(range(2010, 2020))  # Adjust for full range
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Movie', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data (replace with actual full dataset if needed)
df_top10 = pd.DataFrame({
    'movie_title': [
        'Star Wars: The Force Awakens', 'Avengers: Endgame', 'Black Panther',
        'Avengers: Infinity War', 'Jurassic World', 'The Lion King',
        'The Avengers', 'Incredibles 2', 'Beauty and the Beast', 'Frozen'
    ],
    'release_year': [2015, 2019, 2018, 2018, 2015, 2019, 2012, 2018, 2017, 2013],
    'domestic_gross': [936, 858, 700, 679, 652, 543, 623, 609, 504, 477],
    'international_gross': [1131, 1939, 646, 1369, 1019, 1132, 895, 634, 759, 873]
})

# Group total gross by year (domestic only for this plot)
yearly_gross = df_top10.groupby('release_year')['domestic_gross'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_gross, x='release_year', y='domestic_gross', marker='o', color='navy', linewidth=2)

# Annotations for standout movies
highlight_movies = df_top10[df_top10['domestic_gross'] > 850]
for _, row in highlight_movies.iterrows():
    plt.text(row['release_year'], row['domestic_gross'] + 15,
             row['movie_title'], fontsize=9, ha='center', color='darkred', wrap=True)

# Trend line (polynomial regression)
z = np.polyfit(yearly_gross['release_year'], yearly_gross['domestic_gross'], 2)
p = np.poly1d(z)
plt.plot(yearly_gross['release_year'], p(yearly_gross['release_year']), linestyle='--', color='gray', label='Trend')

# Labels, titles, polish
plt.title('Total Domestic Gross by Year (Top 10 Movies, 2010–2018)', fontsize=16)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Domestic Gross (in Millions USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(2010, 2020, 1))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_grossing_movies(df, title_col='movie_title', value_col='domestic_gross'):
    """
    Plot a horizontal bar chart of the top grossing movies.

    Parameters:
    df (DataFrame): DataFrame with movie titles and gross earnings.
    title_col (str): Column name for movie titles.
    value_col (str): Column name for gross earnings.
    """
    # Sort and take top 10
    df_sorted = df.sort_values(by=value_col, ascending=False).head(10)

    # Set figure and plot
    plt.figure(figsize=(12, 6))
    sns.barplot(x=value_col, y=title_col, data=df_sorted,
                palette='Blues_r', edgecolor='black')

    # Add labels and title
    plt.title('Top 10 Grossing Movies (2010–2018)', fontsize=14)
    plt.xlabel('Domestic Gross (in Millions USD)')
    plt.ylabel('Movie Title')
    plt.grid(axis='x', linestyle='--', alpha=0.6)

    # Add value labels
    for index, value in enumerate(df_sorted[value_col]):
        plt.text(value + 5, index, f"${value:.0f}M", va='center', fontsize=9)

    plt.tight_layout()
    plt.show()

# Example usage
plot_top_grossing_movies(df_top10)
```
